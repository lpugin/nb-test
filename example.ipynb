```python
import verovio
import music21
from IPython.display import SVG
```

We need to create a Verovio toolkit instance. We can also set some options, such as a scaling percentage (here 30%)

```python
tk = verovio.toolkit()
tk.setOptions({'scale': 30})
```

```python
example = music21.corpus.parse('bach/bwv355.mxl')
```

```python
GEX = music21.musicxml.m21ToXml.GeneralObjectExporter(example)
musicxmlStr = GEX.parse().decode('utf-8')
```

Now we can load the `musicxmlStr` into the Verovio toolkit instance

```python
tk.loadData(musicxmlStr)
```

We can see how many pages of music we now have in the instance with `getPageCount()`:

```python
tk.getPageCount()
```

Calling `renderToSVG` will return a SVG string for the corresponding page:

```python
svgStr = tk.renderToSVG(1)
```

```python
SVG(svgStr)
```

```python
svgStr = tk.renderToSVG(2)
```

```python
SVG(svgStr)
```
