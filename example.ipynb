```python
import verovio
import music21
```

```python
tk = verovio.toolkit()
tk.setOptions({'scale': 30})
```

```python
example = music21.corpus.parse('bach/bwv350.mxl')
```

```python
GEX = music21.musicxml.m21ToXml.GeneralObjectExporter(example)
musicxmlStr = GEX.parse().decode('utf-8')
```

```python
tk.loadData(musicxmlStr)
```

```python
tk.renderToSVGFile('page01.svg', 1)
```
